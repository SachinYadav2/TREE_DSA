# Level Order Traverse Using Queque Or BFS Seacrch

```python
class Node:
    def __init__(self, d):
        self.data = d
        self.left = None
        self.right = None

def build_tree():
    print("Enter the data:")
    data = int(input())
    
    if data == -1:
        return None
    
    root = Node(data)
    
    print("Enter data for inserting in left of", data)
    root.left = build_tree()
    
    print("Enter data for inserting in right of", data)
    root.right = build_tree()
    
    return root

def level_order_traversal(root):
    if root is None:
        return
    
    queue = [root, None]  # Use a list as a queue, None represents the end of a level
    
    while queue:
        temp = queue.pop(0)
        
        if temp is None:
            # The current level is complete
            print()
            if queue:
                # Queue still has some child nodes, add None to mark the end of the next level
                queue.append(None)
        else:
            print(temp.data, end=" ")
            if temp.left:
                queue.append(temp.left)
            if temp.right:
                queue.append(temp.right)

# Create an empty root node
root = None

# Insert elements into the binary tree
root = build_tree()

# Perform level-order traversal to display the elements
print("Level-Order Traversal:")
level_order_traversal(root)
```

# PostOrder

```python
class Node:
    def __init__(self, d):
        self.data = d
        self.left = None
        self.right = None

def build_tree():
    print("Enter the data:")
    data = int(input())
    
    if data == -1:
        return None
    
    root = Node(data)
    
    print("Enter data for inserting in left of", data)
    root.left = build_tree()
    
    print("Enter data for inserting in right of", data)
    root.right = build_tree()
    
    return root

def level_order_traversal(root):
    if root is None:
        return
    
    queue = [root, None]  # Use a list as a queue, None represents the end of a level
    
    while queue:
        temp = queue.pop(0)
        
        if temp is None:
            # The current level is complete
            print()
            if queue:
                # Queue still has some child nodes, add None to mark the end of the next level
                queue.append(None)
        else:
            print(temp.data, end=" ")
            if temp.left:
                queue.append(temp.left)
            if temp.right:
                queue.append(temp.right)
                
# ______________________--------------------__________________________

def postorder(root):
    if root is None:
        return

    
    
    postorder(root.left)
    postorder(root.right)
    print(root.data, end=" ")


# Create an empty root node
root = None

# Insert elements into the binary tree
root = build_tree()

# Perform level-order traversal to display the elements
print("Level-Order Traversal:")
level_order_traversal(root)

postorder(root)
```

```python
class Node:
    def __init__(self, d):
        self.data = d
        self.left = None
        self.right = None

def build_from_level_order():
    print("Enter data for root:")
    data = int(input())
    if data == -1:
        return None

    root = Node(data)
    queue = [root]

    while queue:
        temp = queue.pop(0)

        print(f"Enter left node for {temp.data}:")
        left_data = int(input())
        if left_data != -1:
            temp.left = Node(left_data)
            queue.append(temp.left)

        print(f"Enter right node for {temp.data}:")
        right_data = int(input())
        if right_data != -1:
            temp.right = Node(right_data)
            queue.append(temp.right)

root = build_from_level_order()
print("Level-Order Traversal:")
level_order_traversal(root)
```
